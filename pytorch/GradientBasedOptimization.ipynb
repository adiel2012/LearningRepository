```python
import torch

dtype = torch.float
device = torch.device("cpu")
```

```python
num_inputs = 5
num_hiddens = 16
num_outputs = 4
num_records = 30
```

```python
x = torch.rand(num_records, num_inputs)
y_hat = torch.rand(num_records, num_outputs)
w1 = torch.rand(num_inputs, num_hiddens, requires_grad=True)
torch.nn.init.xavier_uniform_(w1)
bias1 = torch.rand(num_hiddens, requires_grad=True)
w2 = torch.rand(num_hiddens, num_outputs, requires_grad=True)
torch.nn.init.xavier_uniform_(w2)
bias2 = torch.rand(num_outputs, requires_grad=True)
```

```python
def updateWeight(gradUpdate, params):
    for param in params:
        gradUpdate(param)

def gdCallback(lr):
    def func(w):
        w -= lr * w.grad
    return func

def momentumCallback(momentum = 0.5, n = 1e-3):
    previousDeltas = {}
    def func(w):
        objId = hex(id(w))
        previousDelta = previousDeltas[objId] if objId in previousDeltas.keys() else 0
        delta = -n * w.grad + momentum * previousDelta
        w += delta
        previousDeltas[objId] = delta
    return func

def makeZeroGradient(params):
    for param in params:
        param.grad.zero_()
```

```python
#Training
num_iterations = 500
learning_rate = 1e-3

for i in range(num_iterations):

    #forward
    y = torch.matmul(torch.matmul(x, w1) + bias1, w2) + bias2  

    loss = (y - y_hat).pow(2).sum()
    
    if(i % 1 == 0):
        print(f'Error: {loss.item()}')

    loss.backward()

    with torch.no_grad():
        updateWeight(momentumCallback(), [w2,bias2,w1,bias1])
        makeZeroGradient([w1,w2,bias1,bias2])

print('finish')
```
